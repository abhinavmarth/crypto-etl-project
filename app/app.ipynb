```python
import requests
import mysql.connector
import datetime  

response = requests.get("https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=250&page=1")
data = response.json()
data
```

```python
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="YOUR_PASSWORD",
    database="crypto",
    port=3306
) 

timestamp = datetime.datetime.now()
cur = conn.cursor()
for coin in data:
    cur.execute("""
        INSERT IGNORE INTO coins (name, symbol)
        VALUES (%s, %s)
    """, (coin["name"], coin["symbol"]))
    
    cur.execute("SELECT id FROM coins WHERE name = %s", (coin["name"],))
    coin_id = cur.fetchone()[0]

    cur.execute("""
        INSERT INTO coin_prices (coin_id, price, market_cap, timestamp)
        VALUES (%s, %s, %s, %s)
    """, (coin_id, coin["current_price"], coin["market_cap"], timestamp))
    
conn.commit()
cur.close()
conn.close()
```
